# Pickling all dictionaries of configurations

**Objective -** To determine realistic enough distributions for the parameters ($E_0$, $n_0$, $z$, $\theta_c$, $\theta_w$, $\theta_{obs}) and save the configurations in a Pickle file : 

- 1000 different configurations $ \Longrightarrow $ 1000 dictionaries for 2 values of $\theta_c$ (0.05 and 0.15 radians),
- For a value of $\theta_c$, all the corresponding dictionaries are put in a list $ \Longrightarrow $ 2 lists,
- Each list is saved in a different Pickle file.

**Output :** `all_dico_<thetaCore>.pkl` where `<thetaCore>` is `005` or `015`.

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.patches as mpatches
import afterglowpy as grb
import math 
from astropy.cosmology import Planck18 as cosmo
from scipy.integrate import quad
```

```python
Z = {'jetType':     grb.jet.PowerLaw,     # Jet Type
     'specType':    0,                  # Emission Spectrum
     'b':           4,       # Power Law index
     'thetaObs':    0.2,     # Viewing angle in radians
     'E0':          1.0e51,  # Isotropic-equivalent energy in erg
     'thetaWing':   0.15,    # Truncation angle in radians
     'thetaCore':   0.05,     # Half-opening angle in radians
     'n0':          0.1,     # Circumburst density in cm^{-3}
     'p':           2.2,     # Electron energy distribution index
     'epsilon_e':   0.1,     # epsilon_e
     'epsilon_B':   0.01,    # epsilon_B
     'xi_N':        1.0,     # Fraction of electrons accelerated
     'd_L':         1.0e28,  # Luminosity distance in cm
     'z':           0.5}    # Redshift
```

```python
file1 = open('../data/simulations/all_dico_005.pkl', 'wb')
file2 = open('../data/simulations/all_dico_015.pkl', 'wb')

z = [0.002*i for i in range(1, 51)]
p = [i/1275 for i in range(1, 51)]

all_dico_005 = []
all_dico_015 = []
   
for i in range(1000):
    
    Z['thetaObs'] = np.random.uniform(0, np.pi/2)
    E = np.random.normal(51,1)
    Z['E0'] = 1.0*10**E
    Z['n0'] = 1.0*10**(-np.random.uniform(0, 3))
    Z['z'] = np.random.choice(z, p=p)
    Z['d_L'] = cosmo.luminosity_distance(Z['z']).value*3.08e24
    
    Z['thetaCore'] = 0.05
    Z['thetaWing'] = math.radians(np.exp(np.random.normal(1.94,0.5)))
    while Z['thetaWing'] < Z['thetaCore']:
        Z['thetaWing'] = math.radians(np.exp(np.random.normal(1.94,0.5)))
    all_dico_005.append(Z.copy())
    
    Z['thetaCore'] = 0.15
    while Z['thetaWing'] < Z['thetaCore']:
        Z['thetaWing'] = math.radians(np.exp(np.random.normal(1.94,0.5)))
    all_dico_015.append(Z.copy())

pickle.dump(all_dico_005, file1)
pickle.dump(all_dico_015, file2)

file1.close()
file2.close()
```

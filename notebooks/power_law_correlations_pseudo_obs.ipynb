### Result distributions
#### Mean minimal magnitude

```python
mag_min = []

for i in range(len(lc_open)):
    
    if lc_open[i]==0:
        mag_min.append(np.nan)
        
    else:
        mag = []
        
        for j in range(len(lc_open[i]['mags'])):
            if lc_open[i]['mags'][j] < lc_open[i]['mags_lim'][j]:
                mag.append(lc_open[i]['mags'][j])
                
        if len(mag)==0:
            mag_min.append(np.nan)
            
        else:  
            mag_min.append(min(mag))
```

```python
fig = plt.figure(figsize=(10,8))

plt.hist(mag_min, bins=100, color='skyblue', edgecolor='dodgerblue', alpha=0.5)

plt.title('Minimal magnitude (Power-Law)')
plt.xlabel('minimal magnitude (in days)')
plt.ylabel('number of GRBs')
```

#### Mean time of minimal magnitude

```python
time_max = []

for i in range(len(lc_open)):
    
    if lc_open[i]==0:
        time_max.append(np.nan)
        
    else:
        time_norm = lc_open[i]['time'] - min(lc_open[i]['time'])
        index_max = lc_open[i]['mags'].index(min(lc_open[i]['mags']))
        t_max = time_norm[index_max]

        time_max.append(t_max)
```

```python
fig = plt.figure(figsize=(10,8))

plt.hist(time_max, bins=100, color='skyblue', edgecolor='dodgerblue', alpha=0.5)
#plt.hist(time_max, bins=100, alpha=0.5)

plt.title('Minimal magnitude time (Power-Law)')
plt.xlabel('time since first detection (in days)')
plt.ylabel('number of GRBs')
```

#### $t_{max} - t_0$

```python
Dt = []

for i in range(len(lc_open)):
    
    mag_u, time_u = [], []
    mag_g, time_g = [], []
    mag_r, time_r = [], []
    mag_i, time_i = [], []
    mag_z, time_z = [], []
    mag_y, time_y = [], []
    
    if lc_open[i]==0:
        
        Dt.append(np.nan)
        
    else:
        
        for j in range(len(lc_open[i]['mags'])):
            
            if lc_open[i]['mags'][j] < lc_open[i]['mags_lim'][j]:
                if lc_open[i]['filt'][j]=='b':
                    mag_u.append(lc_open[i]['mags'][j])
                    time_u.append(lc_open[i]['time'][j])
                elif lc_open[i]['filt'][j]=='c':
                    mag_g.append(lc_open[i]['mags'][j])
                    time_g.append(lc_open[i]['time'][j])
                elif lc_open[i]['filt'][j]=='g':
                    mag_r.append(lc_open[i]['mags'][j])
                    time_r.append(lc_open[i]['time'][j])
                elif lc_open[i]['filt'][j]=='orange':
                    mag_i.append(lc_open[i]['mags'][j])
                    time_i.append(lc_open[i]['time'][j])
                elif lc_open[i]['filt'][j]=='r':
                    mag_z.append(lc_open[i]['mags'][j])
                    time_z.append(lc_open[i]['time'][j])
                else:
                    mag_y.append(lc_open[i]['mags'][j])
                    time_y.append(lc_open[i]['time'][j])
                    
        mag_list = [mag_u, mag_g, mag_r, mag_i, mag_z, mag_y]
        time_list = [time_u, time_g, time_r, time_i, time_z, time_y]
        mag_len = [len(i) for i in mag_list]
        
        mags = mag_list[mag_len.index(max(mag_len))]
        time = time_list[mag_len.index(max(mag_len))]
        
        if len(mags)!=0:
            Dt.append(time[mags.index(min(mags))] - time[0])
        else:
            Dt.append(np.nan)
```

```python
fig = plt.figure(figsize=(10,8))

plt.hist(Dt, bins=100, alpha=0.5, color='skyblue', edgecolor='dodgerblue')

plt.title(r'$t_{max} + t_0$ (Power-Law)')
plt.xlabel(r'$t_{max} + t_0$ (in days)')
plt.ylabel('number of GRBs')
```

#### Mean decreasing rate

```python
rate_dec = []
rate_inc = []

for i in range(len(lc_open)):
    
    mag_u, time_u = [], []
    mag_g, time_g = [], []
    mag_r, time_r = [], []
    mag_i, time_i = [], []
    mag_z, time_z = [], []
    mag_y, time_y = [], []
    
    if lc_open[i]==0:
        rate_dec.append(np.nan)
        rate_inc.append(np.nan)
        
    else:
        for j in range(len(lc_open[i]['mags'])):
            if lc_open[i]['mags'][j] < lc_open[i]['mags_lim'][j]:
                if lc_open[i]['filt'][j]=='b':
                    mag_u.append(lc_open[i]['mags'][j])
                    time_u.append(lc_open[i]['time'][j])
                elif lc_open[i]['filt'][j]=='c':
                    mag_g.append(lc_open[i]['mags'][j])
                    time_g.append(lc_open[i]['time'][j])
                elif lc_open[i]['filt'][j]=='g':
                    mag_r.append(lc_open[i]['mags'][j])
                    time_r.append(lc_open[i]['time'][j])
                elif lc_open[i]['filt'][j]=='orange':
                    mag_i.append(lc_open[i]['mags'][j])
                    time_i.append(lc_open[i]['time'][j])
                elif lc_open[i]['filt'][j]=='r':
                    mag_z.append(lc_open[i]['mags'][j])
                    time_z.append(lc_open[i]['time'][j])
                else:
                    mag_y.append(lc_open[i]['mags'][j])
                    time_y.append(lc_open[i]['time'][j])
                    
        mag_list = [mag_u, mag_g, mag_r, mag_i, mag_z, mag_y]
        time_list = [time_u, time_g, time_r, time_i, time_z, time_y]
        mag_len = [len(i) for i in mag_list]
        
        mags = mag_list[mag_len.index(max(mag_len))]
        time = time_list[mag_len.index(max(mag_len))]
    
    
        if len(mags)!=0 and mags.index(min(mags))<mags.index(max(mags)):        
            
            if mags.index(min(mags))==0:
            
                index_min = mags.index(min(mags))
                index_max = mags.index(max(mags))
                rate_dec.append((mags[index_max] - mags[index_min])/(time[index_max] - time[index_min]))
                
                rate_inc.append(np.nan)
                
            else:
                
                index_min = mags.index(min(mags))
                index_max = mags.index(max(mags))
                rate_dec.append((mags[index_max] - mags[index_min])/(time[index_max] - time[index_min]))
                
                mags2 = mags[0:index_min]
                index_max2 = mags.index(max(mags2))
                rate_inc.append((mags[index_min] - mags[index_max2])/(time[index_min] - time[index_max2]))
                
            
        elif len(mags)!=0 and mags.index(min(mags))>mags.index(max(mags)):
            
            if mags.index(min(mags))==(len(mags)-1):
                
                index_min = mags.index(min(mags))
                index_max = mags.index(max(mags))
                rate_inc.append((mags[index_min] - mags[index_max])/(time[index_min] - time[index_max]))
                
                rate_dec.append(np.nan)
                
            else:
                
                index_min = mags.index(min(mags))
                
                index_max = mags.index(max(mags))
                rate_inc.append((mags[index_min] - mags[index_max])/(time[index_min] - time[index_max]))
                
                mags2 = mags[index_min:(len(lc_open[i]['mags'])-1)]
                index_max2 = mags.index(max(mags2))
                rate_dec.append((mags[index_max2] - mags[index_min])/(time[index_max2] - time[index_min]))
                
        else:
            rate_inc.append(np.nan)
            rate_dec.append(np.nan)
```

```python
rate_dec_not_nan = [x for x in rate_dec if not math.isnan(x)]
rate_inc_not_nan = [x for x in rate_inc if not math.isnan(x)]

print('Mean decreasing rate = %.2f mag/days, standard deviation = %.2f' %(np.mean(rate_dec_not_nan), np.std(rate_dec_not_nan)))
print('Mean increasing rate = %.2f mag/days, standard deviation = %.2f' %(np.mean(rate_inc_not_nan), np.std(rate_inc_not_nan)))

fig = plt.figure(figsize=(13,10))

plt.hist(rate_dec, bins=30, label='decreasing rate', color='skyblue', edgecolor='dodgerblue', alpha=0.5)
plt.hist(rate_inc, bins=100, label='increasing rate', color='orange', edgecolor='orangered', alpha=0.5)

plt.title('Decreasing/increasing rate (Power-Law)')
plt.xlabel('rate (in mag/days)')
plt.ylabel('number of GRBs')

plt.axvline(x=np.mean(rate_dec_not_nan), color='blue', linestyle=':', label='mean decreasing rate')
plt.axvline(x=np.mean(rate_inc_not_nan), color='darkorange', linestyle=':', label='mean increasing rate')
plt.legend()
```

#### g-r color

```python
color = []

for i in range(len(lc_open)):
    
    color_inf = []
    color_sup = []
    color_one = []
    
    mag_u, time_u = [], []
    mag_g, time_g = [], []
    mag_r, time_r = [], []
    mag_i, time_i = [], []
    mag_z, time_z = [], []
    mag_y, time_y = [], []
    
    if lc_open[i]==0:
        
        color.append(np.nan)
        
    else:
        
        for j in range(len(lc_open[i]['mags'])):
            
            if lc_open[i]['mags'][j] < lc_open[i]['mags_lim'][j]:
                if lc_open[i]['filt'][j]=='c':
                    mag_g.append(lc_open[i]['mags'][j])
                    time_g.append(lc_open[i]['time'][j])
                elif lc_open[i]['filt'][j]=='g':
                    mag_r.append(lc_open[i]['mags'][j])
                    time_r.append(lc_open[i]['time'][j])
                    
    
        if len(mag_r)!=0 and len(mag_g)!=0:

            for k in range(min(len(mag_r), len(mag_g))):
                if abs(time_g[k] - time_r[k]) < 0.5:

                    if (mag_g[k] - mag_r[k]) < 0.22:
                        color_inf.append(mag_g[k] - mag_r[k])
                        color_one.append((mag_g[k] - mag_r[k]) + (0.22 - np.mean(color_inf)))

                    else:
                        color_sup.append(mag_g[k] - mag_r[k])
                        color_one.append((mag_g[k] - mag_r[k]) + (0.22 - np.mean(color_sup)))
                
            color.append(np.mean(color_one))
            
        else:
            color.append(np.nan)
```

```python
fig = plt.figure(figsize=(10,8))

plt.hist(color, bins=100, color='skyblue', edgecolor='dodgerblue', alpha=0.5)

plt.title('g-r color (Power-Law)')
plt.xlabel('color (in mag)')
plt.ylabel('number of GRBs')
```

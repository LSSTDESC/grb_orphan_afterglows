# Numbers of GRBs
The idea is to try to have some estimate of how many GRBs Rubin could see, starting from GBM GRBs.
This is a *very very rough calculation*, that "fortunately" happens to give reasonable numbers.


## GBM Catalog numbers
https://iopscience.iop.org/article/10.3847/1538-4357/ab7a18

 the 10 year time period from trigger enabling on 2008 July 12 to 2018 July 11. During this time period GBM triggered almost twice a day on transient events, 2356 of which we identified as cosmic GRBs

```python
gbm_per_year = 2356./10.
print(f'GBM sees {gbm_per_year:.1f} GRBs per year in average.')

# gbm non osculted sky
# https://iopscience.iop.org/article/10.3847/1538-4357/835/1/82
gbm_fov = 0.7

# Fermi spends 15% of its time into the SAA
fermi_uptime_saa = 0.85

# expected all GRB detectable by GBM  per year corrected
gbm_per_year_corrected = gbm_per_year/gbm_fov/fermi_uptime_saa
print(f'GBM could see {gbm_per_year_corrected:.1f} GRBs per year in average.')

# on-axis
print(f'\nExpect {gbm_per_year_corrected:.1f} GRBs on-axis with enough energy each year.')

# long vs short
gbm_per_year_corrected_shorts = gbm_per_year_corrected/10
print(f'\nExpect {gbm_per_year_corrected_shorts:.1f} SHORT GRBs on-axis with enough energy each year.')
```

## Total number of GRBs assuming an opening angle

```python
import math
# solid angle for a cone of opening angle theta
def solid(theta):
    return 2*math.pi*(1-math.cos(theta)) 

theta_1 =0.05 # radians
omega_1 = solid(theta_1)
omega_1_ratio = omega_1/4*math.pi

theta_2 =0.15 # radians
omega_2 = solid(theta_2)
omega_2_ratio = omega_2/4*math.pi
```

```python
# total number of short GRBs
gbm_pop_shorts_1 = gbm_per_year_corrected_shorts/omega_1_ratio
print(f'Assuming an opening angle of {theta_1} rad, the expected total population of short GRBs would be {gbm_pop_shorts_1:.1f} each year, ')
gbm_pop_shorts_2 = gbm_per_year_corrected_shorts/omega_2_ratio
print(f'Assuming an opening angle of {theta_2} rad, the expected total population of short GRBs would be {gbm_pop_shorts_2:.1f} each year, ')
```

```python
# counting off-axis in optical, i.e. considering a larger opening angle but not on-axis
# divide by 4 = 2 for half the sky, 2 for nights only
theta_off_1 = 0.5 # 28 degrees
omega_offaxis_1 = solid(theta_off_1) - omega_1
rubin_shorts_offaxis_1 = omega_offaxis_1/4*math.pi * gbm_pop_shorts_1 / 4.

theta_off_2 = 0.5 # 28 degrees
omega_offaxis_2 = solid(theta_off_2) - omega_2
rubin_shorts_offaxis_2 = omega_offaxis_2/4*math.pi * gbm_pop_shorts_2 / 4.

print(f'Assuming an opening angle of {theta_1} rad, and a visibility in the optical up to {theta_off_1} rad, \
we expect {rubin_shorts_offaxis_1:.1f} short GRBs to be observable by Rubin each year')

print(f'Assuming an opening angle of {theta_2} rad, and a visibility in the optical up to {theta_off_2} rad, \
we expect {rubin_shorts_offaxis_2:.1f} short GRBs to be observable by Rubin each year')
```

```python
# fractions
rubin_off_short_frac_1 = rubin_shorts_offaxis_1/gbm_pop_shorts_1
rubin_off_short_frac_2 = rubin_shorts_offaxis_2/gbm_pop_shorts_2
print(f'Fraction of short GRBs off axis but detectable by Rubin is at most from {rubin_off_short_frac_1:.2f} to {rubin_off_short_frac_2:.2f}')
print(f'\nStarting from GBM like population, the fraction of short GRBs off axis but detectable by Rubin is at most 15%')
```

## Rough conclusion
At most 15% of Short GRBs may be "detectable" off-axis by Rubin", that is "consistent" with the effective 5% that comes out of our simulations starting from a chosen population.
A population of 1000 Short GRBs per year is in the right ball park as well, meaning that tenth of OA could be observed each year by Rubin.
